```python
%matplotlib inline
```

```python
# import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot
plt.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label="Boston")
# create labels for axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# set the y limit between 0 and 45
plt.ylim(0, 45)
# create a title
plt.title("PyBer Fare by Month")
# add the legend
plt.legend()
```

```python
# create the plot
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="boston")
# create labels for axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# set the y limit between 0 and 45
plt.ylim(0,45)
# create a title
plt.title("PyBer Fare by Month")
# add a grid
plt.grid()
# add the legend
plt.legend()
```

```python
# create the plot
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label="Boston")
# label axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# title
plt.title("Pyber Fare by Month")
# add a grid
plt.grid()
# set y limits
plt.ylim(0,45)
# add legend
plt.legend()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# create the plot
plt.bar(x_axis, y_axis, color="green", label="Boston")
# label axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# title
plt.title("PyBer Fare by Month")
# add the legend
plt.legend()
```

```python
# create the plot
plt.barh(x_axis, y_axis, color="magenta", label="Boston")

# label axes
plt.xlabel("Fare($)")
plt.ylabel("Date")
# title
plt.title("PyBer Fare by Month")
# add the legend
plt.legend()

plt.gca().invert_yaxis()
```

```python
# create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# create the plot with ax.plot
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# create the plot
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.legend()
plt.xlabel("FARE($)")
plt.ylabel("DATE")
plt.title("PyBer Faare by Month")
```

```python
# create scatterplot
plt.plot(x_axis, y_axis, 'o')
```

```python
# create bubble chart
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
# create plot - object-oriented
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], color="skyblue", alpha=0.2, edgecolors="black", linewidth=2, label="Boston")
ax.legend()
plt.gca().invert_yaxis()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
import statistics
```

```python
# Get the standard deviation of the values in the y-axis
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# object-oriented approach
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# adjust the x-axis
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
# object-oriented approach
fix, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
# add minor ticks

from matplotlib.ticker import MultipleLocator

# increase the size of the plot figure
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# create minor ticks
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
